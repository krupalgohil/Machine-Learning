```python
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
df.head()
```

# 1.Categorical Data


### Countplot


```python
sns.countplot(x=df['Embarked'])
```

### 2.Numerical Data

```python
sns.set_theme(style="whitegrid", context="notebook", palette="Set2")
sns.histplot(x=df['Age'],kde='True')
```

```python
sns.boxplot(x=df['Age'],hue=df['Sex'])
```
