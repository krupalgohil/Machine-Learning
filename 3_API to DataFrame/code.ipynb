```python
import pandas as pd
import requests

# Fetch data from TVMaze API for "breaking bad" shows
url = 'https://api.tvmaze.com/search/shows?q=breaking+bad'
response = requests.get(url)  # Make HTTP GET request to the API
data = pd.DataFrame(response.json())  # Convert JSON response to DataFrame

# Convert DataFrame back to list of dictionaries (needed for json_normalize)
df = data.to_dict('records')

# Flatten nested JSON structure (converts show.id, show.name, etc. into separate columns)
df1 = pd.json_normalize(df)

# Select only the columns we need from the normalized DataFrame
df1 = df1[['score', 'show.id', 'show.url', 'show.name', 'show.type', 
           'show.language', 'show.genres', 'show.status', 'show.rating.average']]

# Rename columns to simpler, more readable names
df1.columns = ['score', 'id', 'url', 'name', 'type', 'language', 'genres', 'status', 'rating_average']

# Export DataFrame to CSV file (without index column)
df1.to_csv("api_fetch_data.csv", index=False)
```
